1) Though Keras/Tensorflow and PyTorch are the most common tools used for training neural networks, scikit-learn has its own class, MLPClassifier, for neural networks, which has the same interface as its other machine learning algorithms.  Try it on the digits dataset below.

```python
from sklearn import datasets
digits = datasets.load_digits()
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, digits.target)
```

```python
# TODO train and score an MLPClassifier
from sklearn.neural_network import MLPClassifier
mlpClass = MLPClassifier()
mlpClass.fit(X_train, y_train)
print(mlpClass.score(X_test, y_test))
```
